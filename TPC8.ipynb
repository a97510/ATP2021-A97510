```python
def getEMD(linha):
    emd=[]
    novalinha=linha.strip("\n")
    campos=novalinha.split(",")
    emd.append("emd"+str(int(campos[1])+1))
    for i in range (2, len(campos)):
        emd.append(campos[i])
    return emd

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        emd=getEMD(linha)
        bd.append(emd)
    return bd
```

```python
BD = []
BD = lerDataset("emd.csv")
print(BD)
```

```python
def chaveOrdem(exame):
    return exame[1] 
    
def listarDataset(bd):
    bd.sort(key=chaveOrdem, reverse=True)
    print("id   data   nome                  apto")
    print("------------------------")
    for emd in bd:
        print(emd[0] + "|" + emd[1] + "|" + emd[3] + " " + emd[2] + "|" + emd[11])

listarDataset(BD)
```

```python
def modalidades(bd):
    l=[]
    for emd in bd:
        if (emd[7] not in l):
            l.append(emd[7])
    l.sort()
    return l 
print(modalidades(BD))
```

```python
def distribPorModalidade(bd):
    distribuicao={}
    for emd in bd:
        if emd[7] in distribuicao.keys():
            distribuicao[emd[7]]=distribuicao[emd[7]] +1
        else:
            distribuicao[emd[7]]=1   

    return distribuicao
print(distribPorModalidade(BD))
```

```python
def distribPorClube(bd):
    distribuicao={}
    for emd in bd:
        if emd[8] in distribuicao.keys():
            distribuicao[emd[8]]=distribuicao[emd[8]] +1
        else:
            distribuicao[emd[8]]=1   

    return distribuicao
print(distribPorClube(BD))
```

```python
def distribPorAno(bd):
    distribuicao={}
    for emd in bd:
        if emd[1][0:4] in distribuicao.keys():
            distribuicao[emd[1][0:4]]=distribuicao[emd[1][0:4]] +1
        else:
            distribuicao[emd[1][0:4]]=1   

    return distribuicao
distribPorAno(BD)
```

```python
import matplotlib.pyplot as plt
def plotdistribPorModalidade(bd):
    c = distribPorModalidade(bd)
    x = c.keys()
    y = []
    for i in x:
        y.append(c[i])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.bar(x, y)
    plt.show()
plotdistribPorModalidade(BD)
```

```python
import matplotlib.pyplot as plt
def plotdistribPorAno(bd):
    c = distribPorAno(bd)
    x = c.keys()
    y = []
    for i in x:
        y.append(c[i])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.bar(x, y)
    plt.show()
plotdistribPorAno(BD)
```

```python
import matplotlib.pyplot as plt
def plotdistribPorClube(bd):
    c = distribPorClube(bd)
    x = c.keys()
    y = []
    for i in x:
        y.append(c[i])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.bar(x, y)
    plt.show()
plotdistribPorClube(BD)
```

```python
import PySimpleGUI as sg

menu_list_column = [
    [sg.Button("Carregar BD")],
    [sg.Button("Listar BD")],
    [sg.Button("Modalidades")],
    [sg.Button("Distribuição por ano")],
    [sg.Button("Distribuição por clube")],
    [sg.Button("Distribuição por modalidade")],    
]

data_viewer_column = [
    [sg.Text("Painel de Dados")],
    [sg.Text(size=(100, 3), key="-Dados-")],
]


layout = [
    [
        sg.Column(menu_list_column),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]

window = sg.Window("Sample App", layout)


stop = False
while not stop:
    event, values = window.read()
    BD = lerDataset("emd.csv")
    if event == "Sair" or event == sg.WIN_CLOSED:
        stop = True
    elif event == "Carregar BD": 
        window["-Dados-"].update("Foram carregados " + str(len(BD)) + " registos...")
    elif event == "Listar BD":
        myBD = listarDataset(BD)
        window["-Dados-"].update("Os registos carregados foram:")
        window["-Dados-"].update(myBD)
    elif event == "Modalidades":
        myBD = modalidades(BD)
        window["-Dados-"].update("Vou listar que modalidades existem...")
        window["-Dados-"].update(myBD)
    elif event == "Distribuição por ano":
        myBD = distribPorAno(BD)
        myBD2 = plotdistribPorAno(BD)
        window["-Dados-"].update("Vou fazer a distribuição por ano...")
        window["-Dados-"].update(myBD)
        window["-Dados-"].update(myBD2)
    elif event == "Distribuição por clube": 
        myBD = distribPorClube(BD) 
        myBD2 = plotdistribPorClube(BD)
        window["-Dados-"].update("Vou fazer a distribuição por clube...")
        window["-Dados-"].update(myBD) 
        window["-Dados-"].update(myBD2)
    elif event == "Distribuição por modalidade": 
        myBD = distribPorModalidade(BD)
        myBD2 = plotdistribPorModalidade(BD)
        window["-Dados-"].update("Vou fazer a distribuição por modalidade...")
        window["-Dados-"].update(myBD)
        window["-Dados-"].update(myBD2)
    else:
        window["-Dados-"].update("Erro: evento desconhecido.")

window.close()
```
