```python
import matplotlib.pyplot as plt

def menu():
    
    Pol = []

    print ("""Aplicação para trabalhar com polinómios
1) Criar Polinómio
2) Calcular Polinómio
3) Calcular Tabela
4) Simplificar Polinómio
5) Derivar Polinómio
6) Representação gráfica do polinómio
0) Sair""")
    a = int(input("Que opção deseja escolher?"))
    
    while a != 0: 
        if a == 1:
            Pol = criarPolinomio()
            n = int(input("Quantos termos tem o polinómio? "))
            for i in range(n):
                c = int(input("Coeficiente do termo" + str(i+1) + ":"))
                e = int(input("Expoente do termo" + str(i+1) + ":"))
                inserirTermo(Pol, (c,e))
            verPolinomio(Pol)
            print(f"O polinómio criado é:{Pol} ")
            
            a = int(input("Que opção deseja escolher?"))
            
        elif len(Pol) == 0:
            print("Ainda não foi criado um polinómio!")
            a = int(input("Que opção deseja escolher?"))
  
        else:
            if a == 0:
                print("Até à próxima!")
                
            elif a == 2:
                b = int(input("Por que termo desejaria substituir o polinómio?"))
                print(f"{calcPol(Pol, b)}")
                
            elif a == 3:
                b = int(input("Quer que a tabela tenha quantas linhas?"))
                print(tabela(Pol, b))
                
            elif a == 4:
                ordenaPol(Pol)
                print(simplificarPol(Pol))
            
            elif a == 5:
                print(derivarPol(Pol))
                
            elif a == 6:
                grafico(Pol)
            
            a = int(input("Que opção deseja escolher?"))
 

def criarTermo(coef, exp):
    return(coef, exp)


def criarPolinomio():
    return []


def inserirTermo(p, t):
    return p.append(t)


def verTermo(t):
    c,e = t
    if e == 0:
        return str(c)
    elif e == 1:
        return str(c) +'x'
    else:
        return (str(c) +"x^"+ str(e))


def verPolinomio(p):
    res = ""
    for t in p:
        if res == "":
            res = verTermo(t)
        else:
            res = res + "+" + verTermo(t)
    print (res)
            

def calcPol(p, x):
    res = 0
    for (c, e) in p:
        res = res + c*(x**e)
    return res
    
    
def tabela(p, nlinhas):
    for i in range (nlinhas):
        print(str(i) + "::" + str(calcPol(p,i)))


def chaveOrd(t):
    c, e = t
    return e


def ordenaPol(p):
    p.sort(reverse = True, key = chaveOrd)


def simplificarPol(p):
    ordenaPol(p)
    poln = []
    for i in range (1, len(p)):
        coef,exp =p [i]
        coef0,exp0 = p[i-1]
        if exp0 == exp:
            inserirTermo(poln, (coef+coef0, exp))
        else:
            if exp0 > exp:
                if i == 1:
                    inserirTermo(poln, (coef0, exp0))
            else:
                inserirTermo(poln, (coef, exp))
    return verPolinomio(poln)


def derivarPol(p):
    res = []
    for (c, e) in p:
        if e > 0:
            res.append((c*e, e-1))
    return res


def grafico(myPol):
    x = [i for i in range (-10,10)]
    y = []
    for n in x:
        y.append(calcPol(myPol,n))

    plt.xlabel("x")
    plt.ylabel("p(x)")
    plt.plot(x, y, label = "polinómio")

    eixoXx = [-10, 10]
    eixoXy = [0, 0]
    plt.plot(eixoXx, eixoXy, label = "eixo dos X")

    eixoYx = [0, 0]
    eixoYy = [-4000, 6000]
    plt.plot(eixoYx, eixoYy, label = "eixo dos Y")

    plt.title("Polinómios")
    plt.legend()
    plt.show()


menu()
```
