```python
import io
import matplotlib.pyplot as plt

def getAluno(texto):
    texto2 = texto.replace("\"", "")
    campos = texto2.split(",")
    aluno = []
    aluno.append(campos[0])
    aluno.append(campos[1])
    aluno.append(campos[2])
    aluno.append(campos[3:])
    lista = []
    for s in aluno[3]:
        lista.append(int(s))
    aluno[3] = lista
    return aluno


def lerDataset(fnome):
    f = io.open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        bd.append(getAluno(linha))
    return bd


def chaveOrd(a):
    return a[1]


def listarDataset(bd):
    bd.sort(key = chaveOrd)
    print("id          nome          curso          média")
    print("----------------------------------------------")
    for a in bd:
        print(a[0] + " | " + a[1] + " | " + a[2] + " | " + str(sum(a[3])/4))


def consultarDataset(bd, id):
    for i in range(0,len(bd)):
        if bd[i][0] == id:
            med = media(bd[i][3])
            return print("INFO :: " + str(bd[i]) + "\nMEDIA :: " + str(med))
        else:
            if i == len(bd)-1:
                return print('Não existe nenhum aluno com esse id!')
            

def chaveOrdena(a):
    return media(a[3])


def top10(bd):      
    bd.sort(key=chaveOrdena, reverse= True)
    novabd = []             
    for aluno in range(0,10):
        novoAluno = []
        for parametro in range(0,3):           
            novoAluno.append(bd[aluno][parametro])
        novoAluno.append(media(bd[aluno][3]))  
        novabd.append(novoAluno)
    return novabd


def distribuirPorCurso(bd):
    distribuicao = {}
    for a in bd:
        if a[2] in distribuicao.keys():
            distribuicao[a[2]] = distribuicao[a[2]] + 1
        else:
            distribuicao[a[2]] = 1
    return distribuicao


def media(l):
    soma = 0
    for i in range(0, len(l)):
        soma = soma + l[i]
    return soma/4


def distribuirPorMedia(bd):
    distribuicao = {}
    for a in bd:
        if str(round(media(a[3]))) in distribuicao.keys(): 
            distribuicao[str(round(media(a[3])))] = distribuicao[str(round(media(a[3])))] + 1
        else:
            distribuicao[str(round(media(a[3])))] = 1
    return distribuicao


def graficoDistribuirPorCurso(bd):
    c = distribuirPorCurso(bd)
    x = c.keys()
    y = []
    for i in x:
        y.append(c[i])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.bar(x, y)
    plt.show()
    

def graficoDistribuirPorMedia(bd):
    c = distribuirPorMedia(bd)
    x = c.keys()
    y = []
    for i in x:
        y.append(c[i])
    plt.xlabel("x")
    plt.ylabel("y")
    plt.bar(x, y)
    plt.show()


bd = lerDataset("alunos.csv")
print(bd)
listarDataset(bd)
id = input("Qual o id que deseja procurar?")
consultarDataset(bd, id)
for linha in top10(bd):
    print(linha)
print(distribuirPorCurso(bd))
print(distribuirPorMedia(bd))
graficoDistribuirPorCurso(bd)
graficoDistribuirPorMedia(bd)
```
